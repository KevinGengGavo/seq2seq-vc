```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
import librosa
```

```python
x = "/home/kevingenghaopeng/vc/seq2seq-vc/egs/arctic/vc2/downloads/V006_S1_max_valid/0668-0004.wav"
y = "/home/kevingenghaopeng/vc/seq2seq-vc/egs/arctic/vc2/downloads/V006_SS_max_valid/0668-0004.wav"
z = "/home/kevingenghaopeng/vc/seq2seq-vc/egs/arctic/vc2/downloads/V000_R_max_valid/0668-0004.wav"
```

```python
s1, sr = librosa.load(x)
ss, sr = librosa.load(y)
r, sr = librosa.load(z)
# normalize
s1 = s1 / max(abs(s1))
ss = ss / max(abs(ss))
r = r / max(abs(r))
```

```python
# plot y 
color_map = {"R": '#6A7D8A', "SS": '#6B8E23', "S1": '#D2691E'}
for voice_type, x in {"S1": s1, "SS": ss, "R": r}.items():
    figsize = (12, 3)
    fig = plt.figure(figsize=figsize, dpi=500)
    ax = fig.add_subplot(111)
    ax.plot(x, linewidth=0.5, color=color_map[voice_type])
    # hide x label and y lable
    custom_lines = [Line2D([0], [0], color=color_map[voice_type], lw=4)]
    legend = ax.legend(custom_lines, [voice_type], loc='upper right', fontsize=15)
    ax.set_xticks([])
    ax.set_yticks([])
    # show figure
    fig.show()
    fig.tight_layout()
    fig.savefig(voice_type + ".png")
```
